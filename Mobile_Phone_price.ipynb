```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop(['price_range'],axis=1)
```

```python
y=data['price_range']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=42)
```

```python
print(X_train)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
model1=RandomForestClassifier(random_state=20)
model1.fit(X_train,y_train)
```

```python
predict_y=model1.predict(X_test)
accuracy=accuracy_score(y_test,predict_y)
```

```python
print(f"RandomForestClassifier Accuracy:{accuracy*100}")
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
gnb.fit(X_train,y_train)
```

```python
predict_y_gnb=gnb.predict(X_test)
```

```python
accuracy_gnb=accuracy_score(y_test,predict_y_gnb)
```

```python
print(f"gaussian Naive Bayes Accuracy:{accuracy_gnb*100}")
```
